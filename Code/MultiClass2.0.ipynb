```python
def Model_OneVsRest_Hier_notShared(nb_n1=100,nb_n2=50):
    #l'apprentissage de chaque partie est independant, c'est du vrai multi task
    inputs = Input(shape=(600,))
    #Partie Random
    x_random=Dense(nb_n1, activation='sigmoid')(inputs)
    out_pre_random=Dense(nb_n2, activation='sigmoid')(x_random)
    out_random=Dense(1, activation='sigmoid')(out_pre_random)
    
  

    inputs_mero_W=Lambda(lambda x: x*(1-out_random))(inputs)
    x_mero=Dense(nb_n1, activation='sigmoid')(inputs_mero_W)
    out_pre_mero=Dense(nb_n2, activation='sigmoid')(x_mero)
    out_mero=Dense(1, activation='sigmoid')(out_pre_mero)
    
    
    inputs_hyper_W=Lambda(lambda x: x*(1-out_mero))(inputs)
    x_hyper=Dense(nb_n1, activation='sigmoid')(inputs_hyper_W)
    out_pre_hyper=Dense(nb_n2, activation='sigmoid')(x_hyper)
    out_hyper=Dense(1, activation='sigmoid')(out_pre_hyper)
    
    
    inputs_coh_W=Lambda(lambda x: x*(1-out_hyper))(inputs)
    x_coh=Dense(nb_n1, activation='sigmoid')(inputs_coh_W)
    out_pre_coh=Dense(nb_n2, activation='sigmoid')(x_coh)
    out_coh=Dense(1, activation='sigmoid')(out_pre_coh)
    
    inputs_syn_W=Lambda(lambda x: x*(1-out_coh))(inputs)
    x_syn=Dense(nb_n1, activation='sigmoid')(inputs_syn_W)
    out_pre_syn=Dense(nb_n2, activation='sigmoid')(x_syn)
    out_syn=Dense(1, activation='sigmoid')(out_pre_syn)
    

    loss_f='binary_crossentropy'
    model = Model(inputs=[inputs], outputs=[out_coh,out_hyper,out_mero,out_random,out_syn])
    model.compile(optimizer='Adam', loss= [loss_f,loss_f,loss_f,loss_f,loss_f])
    return model
```

```python
def Model_OneVsRest_Hier3(nb_n1=100,nb_n2=50):
    #l'apprentissage de chaque partie est independant, c'est du vrai multi task
    inputs = Input(shape=(600,))
    #Partie Related_or Not
    x_random=Dense(nb_n1, activation='sigmoid')(inputs)
    out_pre_random=Dense(nb_n2, activation='sigmoid')(x_random)
    out_random1=Dense(1, activation='sigmoid')(out_pre_random)

    
    inputs_mero=concatenate([inputs,x_random])
    inputs_mero_W=Lambda(lambda x: 2*x-x*(out_random1))(inputs_mero)
    x_mero=Dense(nb_n1, activation='sigmoid')(inputs_mero_W)
    out_pre_mero=Dense(nb_n2, activation='sigmoid')(x_mero)
    out_mero1=Dense(1, activation='sigmoid')(out_pre_mero)
  
    
    inputs_hyper=concatenate([inputs,x_mero])
    inputs_hyper_W=Lambda(lambda x: 2*x-x*(out_random1*out_mero1))(inputs_hyper)
    x_hyper=Dense(nb_n1, activation='sigmoid')(inputs_hyper_W)
    out_pre_hyper=Dense(nb_n2, activation='sigmoid')(x_hyper)
    out_hyper1=Dense(1, activation='sigmoid')(out_pre_hyper)

    
    inputs_coh=concatenate([inputs,x_hyper])
    inputs_coh_W=Lambda(lambda x: 2*x-x*(out_random1*out_mero1*out_hyper1))(inputs_coh)
    x_coh=Dense(nb_n1, activation='sigmoid')(inputs_coh_W)
    out_pre_coh=Dense(nb_n2, activation='sigmoid')(x_coh)
    out_coh1=Dense(1, activation='sigmoid')(out_pre_coh)

    
    inputs_syn=concatenate([inputs,x_coh])
    inputs_syn_W=Lambda(lambda x: 2*x-x*(out_random1*out_mero1*out_hyper1*out_coh1))(inputs_syn)
    x_syn=Dense(nb_n1, activation='sigmoid')(inputs_syn)
    out_pre_syn=Dense(nb_n2, activation='sigmoid')(x_syn)
    out_syn1=Dense(1, activation='sigmoid')(out_pre_syn)
  
    out_random_pref=concatenate([out_random1,out_mero1,out_hyper1,out_coh1,out_syn1])
    
    #out_random=Dense(1, activation='sigmoid')(out_random_pref)
    out_random=out_random1

    loss_f='binary_crossentropy'
    model = Model(inputs=[inputs], outputs=[out_coh1,out_hyper1,out_mero1,out_random,out_syn1])
    model.compile(optimizer='Adam', loss= [loss_f,loss_f,loss_f,loss_f,loss_f])
    return model
```

```python
def train_model_softmax(name_model,epochs):
    model=name_model
    
    model.fit(xtrain, [ytrain.transpose()[0],ytrain.transpose()[1],ytrain.transpose()[2]
                       ,ytrain.transpose()[3],ytrain.transpose()[4]],
                epochs=epochs, validation_data=(xvalid, [yvalid.transpose()[0],yvalid.transpose()[1],yvalid.transpose()[2]
                       ,yvalid.transpose()[3],yvalid.transpose()[4]]),  verbose=False, callbacks=[EarlyStopping(patience=5)])
   
    return model
```

```python
def func_preds_max_softmax(model,xtest):
    preds = model.predict(xtest, verbose=False)
    preds_pre=[]

    for i in range(len(preds[0])):
        a=np.zeros(5)
        b=[]
        for j in range(5):
            b.append(preds[j][i])
        a[np.argmax(b)]=1
        preds_pre.append(a)
    preds_all_max=[list_cat[np.argmax(l)] for l in preds_pre]

    return preds_all_max
    
```

```python

def create_metrics(y_test,y_pred):
    y_test=[list_cat[np.argmax(l)] for l in y_test]
    cm = ConfusionMatrix(actual_vector=y_test, predict_vector=y_pred) # Create CM From Data
    dict_F1=cm.class_stat['F1']
    df_metrics=pd.DataFrame.from_dict(dict_F1, orient='index',columns=['F1'])
    df_metrics['ACC']=pd.DataFrame.from_dict(cm.class_stat['ACC'], orient='index' )

    #cm.class_stat
    return df_metrics
```

```python
def main_softmax(name_model,xtest,epochs):
    model=train_model_softmax(name_model,epochs)
    preds_all_max=func_preds_max_softmax(model,xtest)
    df_metrics_max=create_metrics(ytest,preds_all_max)
    return df_metrics_max
```

```python
def Model_OneVsRest_Hier_8(nb_n1=100,nb_n2=50):
    #l'apprentissage de chaque partie est independant, c'est du vrai multi task
    inputs = Input(shape=(600,))
    #Partie Random
    x_random=Dense(nb_n1, activation='sigmoid')(inputs)
    out_pre_random=Dense(nb_n2, activation='sigmoid')(x_random)
    out_r=Dense(2, activation='softmax')(out_pre_random)
    
    out_random1=Lambda(lambda x: x[:,0:1])(out_r)
    out_not_r=Lambda(lambda x: x[:,1:])(out_r)

    
    inputs_mero=concatenate([inputs,x_random])
    inputs_mero_W=Lambda(lambda x: x*(out_not_r+0.1))(inputs_mero)
    x_mero=Dense(nb_n1, activation='sigmoid')(inputs_mero_W)
    out_pre_mero=Dense(nb_n2, activation='sigmoid')(x_mero)
    out_mero1=Dense(2, activation='softmax')(out_pre_mero)
    
    out_mero=out_random1=Lambda(lambda x: x[:,0:1])(out_mero1)
    out_not_m=Lambda(lambda x: x[:,1:])(out_mero1)
    
    
    
    inputs_hyper=concatenate([inputs,x_mero])
    inputs_hyper_W=Lambda(lambda x: x*(minimum([out_not_m,out_not_r])+0.1))(inputs_hyper)
    x_hyper=Dense(nb_n1, activation='sigmoid')(inputs_hyper_W)
    out_pre_hyper=Dense(nb_n2, activation='sigmoid')(x_hyper)
    out_hyper1=Dense(2, activation='softmax')(out_pre_hyper)
    
    out_hyper=Lambda(lambda x: x[:,0:1])(out_hyper1)
    out_not_hy=Lambda(lambda x: x[:,1:])(out_hyper1)
    
    
    
    inputs_coh=concatenate([inputs,x_hyper])
    inputs_coh_W=Lambda(lambda x:  x*(minimum([out_not_m,out_not_r,out_not_hy])+0.1))(inputs_coh)
    x_coh=Dense(nb_n1, activation='sigmoid')(inputs_coh_W)
    out_pre_coh=Dense(nb_n2, activation='sigmoid')(x_coh)
    out_coh1=Dense(2, activation='softmax')(out_pre_coh)
    
    out_coh=Lambda(lambda x: x[:,0:1])(out_coh1)
    out_not_c=Lambda(lambda x: x[:,1:])(out_coh1)
    
    
    inputs_syn=concatenate([inputs,x_coh])
    inputs_syn_W=Lambda(lambda x: x*(minimum([out_not_m,out_not_r,out_not_hy,out_not_c])+0.1))(inputs_syn)
    x_syn=Dense(nb_n1, activation='sigmoid')(inputs_syn_W)
    out_pre_syn=Dense(nb_n2, activation='sigmoid')(x_syn)
    out_syn1=Dense(2, activation='softmax')(out_pre_syn)
    
    out_syn=Lambda(lambda x: x[:,0:1])(out_syn1)
    out_not_s=Lambda(lambda x: x[:,1:])(out_syn1)
    
    out_random=out_random1
  

    loss_f='binary_crossentropy'
    model = Model(inputs=[inputs], outputs=[out_coh,out_hyper,out_mero,out_random,out_syn])
    model.compile(optimizer='Adam', loss= [loss_f,loss_f,loss_f,loss_f,loss_f])
    return model
```
